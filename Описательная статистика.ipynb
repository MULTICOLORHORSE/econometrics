```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel(r"C:\Users\Igor\Downloads\Dataset.xlsx")
```

```python
def desc(X):
    desc_d = {}
    desc_d.update({'NAME':X.name})
    desc_d.update({'MEAN':np.mean(X)})
    desc_d.update({'MODE':stats.mode(X).mode[0]})
    desc_d.update({'MODE_COUNT':stats.mode(X).count[0]})
    desc_d.update({'MEDIAN':np.median(X)})  
    desc_d.update({'VAR':np.var(X)})
    desc_d.update({'STD':np.std(X)})
    desc_d.update({'MIN': min(X)})
    desc_d.update({'MAX':max(X)})
    desc_d.update({'IQR':stats.iqr(X)})
    desc_d.update({'SKEW':stats.skew(X)})
    desc_d.update({'KURT':stats.kurtosis(X)})
    return(desc_d)
```

```python
base = []
for col in data.columns[3:]:
    base.append(desc(data[col]))
```
